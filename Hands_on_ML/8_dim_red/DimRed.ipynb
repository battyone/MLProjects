```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.decomposition import KernelPCA
from sklearn.manifold import TSNE
```

```python
from sklearn.datasets import load_iris, load_breast_cancer
```

```python
data_br = load_breast_cancer()
```

```python
data_br.data
```

```python
for i in [20, 25, 30, 35, 40]:
    for _ in range(1):
        X_embedded = TSNE(n_components=2, verbose=0, perplexity=i).fit_transform(data_br.data)
        df = pd.DataFrame(X_embedded)
        plt.scatter(df[0], df[1], c = data_br.target)
    plt.show()
```

```python
for i in [10**-10, 0.001, 0.1, 1, 10]:
    rbf_pca = KernelPCA(n_components=2, kernel='rbf', gamma=i, alpha=100)
    X_red = rbf_pca.fit_transform(data_br.data)
    df_iris = pd.DataFrame(X_red)
    plt.scatter(df_iris[0], df_iris[1], c = data_br.target)
    #plt.scatter(df_iris[0], df_iris[1])
    plt.show()
```

```python
from sklearn.manifold import LocallyLinearEmbedding
```

```python
for i in [40, 60, 80, 100]:
    lle = LocallyLinearEmbedding(n_components=2, n_neighbors=i)
    x_red_lle = lle.fit_transform(data_br.data)
    df_red_lle = pd.DataFrame(x_red_lle)
    plt.scatter(df_red_lle[0], df_red_lle[1], c = data_br.target)
    plt.show()
```

```python
from sklearn.manifold import Isomap, MDS, SpectralEmbedding
```

```python
for i in [1, 5, 10, 20, 30, 70]:
    ism = Isomap(n_components=2, n_neighbors=i)
    x_red_ism = ism.fit_transform(data_br.data)
    df = pd.DataFrame(x_red_ism)
    plt.scatter(df[0], df[1], c = data_br.target)
    plt.show()
```

```python
for i in [10**-5, 10**-1, 10**0, 10**5]:
    model = MDS(n_components=2,n_init=10, eps=i)
    X = model.fit_transform(data_br.data)
    df = pd.DataFrame(X)
    plt.scatter(df[0], df[1], c = data_br.target)
    #plt.scatter(df[0], df[1])
    plt.show()
```

```python
for _ in range(3):
    model = SpectralEmbedding(n_components=2, affinity='rbf')
    X = model.fit_transform(data_br.data)
    df = pd.DataFrame(X)
    plt.scatter(df[0], df[1], c = data_br.target)
    #plt.scatter(df[0], df[1])
    plt.show()
```

```python
for _ in range(3):
    model = KernelPCA(n_components=12, kernel='rbf', gamma=i, alpha=100)
    X = model.fit_transform(data_br.data)
    model = KernelPCA(n_components=10, kernel='rbf', gamma=10**-8, alpha=100)
    X = model.fit_transform(X)
    model = LocallyLinearEmbedding(n_components=8, n_neighbors=50)
    X = model.fit_transform(X)
    model = Isomap(n_components=6, n_neighbors=10)
    X = model.fit_transform(X)
    model = MDS(n_components=4,n_init=10, eps=10**-1)
    X = model.fit_transform(X)
    X = TSNE(n_components=2, verbose=0, perplexity=10).fit_transform(X)

    #model = SpectralEmbedding(n_components=2, affinity='rbf')
    #X = model.fit_transform(data_br.data)
    
    df = pd.DataFrame(X)
    plt.scatter(df[0], df[1], c = data_br.target)
    plt.show()
```

```python
for _ in range(3):
    model = KernelPCA(n_components=6, kernel='rbf')
    X = model.fit_transform(data_br.data)
    model = LocallyLinearEmbedding(n_components=5, n_neighbors=50)
    X = model.fit_transform(X)
    model = Isomap(n_components=4, n_neighbors=10)
    X = model.fit_transform(X)
    model = MDS(n_components=3,n_init=10, eps=10**-1)
    X = model.fit_transform(X)
    X = TSNE(n_components=2, verbose=0, perplexity=30).fit_transform(X)

    df = pd.DataFrame(X)
    plt.scatter(df[0], df[1], c = data_br.target)
    plt.show()
```

```python
for _ in range(3):
    model = LocallyLinearEmbedding(n_components=6, n_neighbors=50)
    X = model.fit_transform(data_br.data)
    
    model = KernelPCA(n_components=5, kernel='rbf')
    X = model.fit_transform(X)
    
    model = Isomap(n_components=4, n_neighbors=10)
    X = model.fit_transform(X)
    
    model = MDS(n_components=3,n_init=10, eps=10**-1)
    X = model.fit_transform(X)
    
    X = TSNE(n_components=2, verbose=0, perplexity=50).fit_transform(X)

    df = pd.DataFrame(X)
    plt.scatter(df[0], df[1], c = data_br.target)
    plt.show()
```
